```python
y = data['amount_of_rain'].to_numpy()
x = data['temp'].to_numpy()
```

```python
x = np.insert([x],0,[np.ones(x.shape[0])],axis=0)
```

```python
x = x.T
```

```python
theta = np.linalg.inv(x.T@x)@x.T@y
```

```python
import matplotlib.pyplot as plt
```

```python
theta
```

```python
plt.plot(x[:,1],y,'bo',label="True Rain Fall")
plt.plot(x[:,1],x@theta,'b',label="predict")
plt.title("Correlation between rain fall and temp")
plt.xlabel('temperature')
plt.ylabel('rain fall')
plt.legend()
plt.show()
```
